# Programming Book

Buku merupakan salah satu sumber informasi yang sering kali digunakan untuk menambah pengetahuan banyak orang. Terdapat berbagai macam buku, mulai dari buku sejarah, buku fiksi hingga buku mengenai programming. Bagi seorang programmer, membaca buku programming merupakan salah satu hal yang perlu dilakukan untuk mendapatkan informasi mengenai programming serta juga mendapatkan insight baru terkait dengan penggunaan sebuah bahasa pemrograman. Namun, bagaimana cara kita mengetahui bahwa apakah buku programming yang kita baca itu bagus?

Dataset prog_book.csv merupakan dataset yang berisikan tentang 270 programming book. Terdapat 7 kolom pada dataset tersebut yang terdiri dari:
1. Rating = Merupakan rating dari buku programming tersebut
2. Reviews = Jumlah orang yang memberikan review terkait dengan buku programming
3. Book_title = Judul dari buku tersebut
4. Description = Merupakan deskripsi singkat dari buku
5. Number_Of_Pages = Jumlah halaman dari buku
6. Type = Jenis buku programming
7. Price = Harga(Dalam satuan USD)

#### Problem: Apa kriteria untuk buku  programming yang bagus?

Terlebih dahulu data dibaca

```python
import pandas as pd

df = pd.read_csv("prog_book.csv")
```

```python
df
```

Melakukan pemeriksaan terhadap data type dari setiap kolom, dilihat bahwa reviews memiliki type data object sehingga nanti perlu diubah

```python
df.info()
```

Melakukan pemeriksaan terhadap missing value, tidak terdapat missing value pada dataset tersebut

```python
df.isna().sum()
```

Menghilangkan tanda koma pada kolom reviews agar tipe data nya bisa dirubah

```python
df['Reviews'] = df['Reviews'].str.replace(',','')
```

Mengubah tipe datanya menjadi numeric

```python
df['Reviews'] = pd.to_numeric(df['Reviews'],errors='coerce')
```

```python
df.info()
```

```python
df
```

Dapat dilihat bahwa, nilai pada reviews adalah 0 yang artinya tidak ada yang me review namun terdapat nilai rating pada kolom rating. Sehingga terdapat kebingungan dalam data tersebut, sehingga sebaiknya kita menghapus data yang memiliki review 0

```python
reviews_0 = df[df['Reviews'] == 0]
reviews_0
```

```python
df.drop(reviews_0.index,inplace=True)
```

```python
df
```

```python
df[df['Reviews'] == 0]
```

Setelah data yang kita miliki sudah "bersih", kita bisa melakukan exploratory data analysis terkait dengan pertanyaan yang kita miliki. Pertama, kita perlu melihat dari rating buku programming, bahwa berapa minimum rating yang diperlukan agar sebuah buku dikategorikan buku programming yang bagus

Untuk menentukannya, kita menggunakan grafik histogram untuk melihat distribusi data dari rating  buku programming

```python
df.info()
```

```python
df['Rating']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(df,x='Rating')
plt.title("Distribution of Programming Book Rating")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

Dapat dilihat pada grafik tersebut bahwa, buku programming yang bagus perlu memiliki rating lebih dari 4.2

Setelah mendapatkan data tersebut, kira perlu memeriksa bahwa berapa banyak buku yang memiliki rating lebih dari 4.2

```python
good_rating  = df[df['Rating'] > 4.2]
good_rating
```

Dapat dilihat bahwa terdapat 84 data yang ada, sehingga selanjutnya kita akan menggunakan 84 data tersebut

Selanjutnya kita akan mencari buku bertipe apa yang memiliki rata rata paling tinggi diantara tipe lainnya

```python
book_type = good_rating.groupby('Type')['Rating'].mean()
book_type.plot(kind='bar',figsize=(10,4),rot=0)
plt.title("Average rating for each book type")
plt.xlabel("Type")
plt.ylabel("Rating")
plt.show()
```

Dapat dilihat bahwa tipe buku dengan rating tertinggi adalah Boxed Set - Hardcover(Namun untuk tipe sendiri opsional, jadi ini tidak bisa dijadikan patokan)

Setelah itu, kita perlu melihat berapa kira kira jumlah halaman untuk membuat sebuah buku programming tersebut dikategorikan bagus

```python
plt.hist(good_rating['Number_Of_Pages'])
plt.title("Distribution of Programming Book Number of Pages")
plt.xlabel("Number of Pages")
plt.ylabel("Count")
plt.show()
```

Dapat dilihat bahwa rata rata jumlah halaman pada buku programming yang memiliki rating diatas 4.2 adalah kurang dari 500 halaman

Setelah itu, kita perlu melihat berapa kira kira harga yang ditetapkan pada buku programming yang bagus

```python
plt.hist(good_rating['Price'])
plt.title("Distribution of Programming Book Price")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

Dapat dilihat bahwa rata rata harga untuk buku programming yang bagus diantara 40-50 USD

#### Kesimpulan

Sehingga dapat disimpulkan bahwa buku programming yang bagus adalah yang memiliki rating diatas 4.2 dengan jumlah halaman kurang dari 500 dan kisaran harganya di 40-50 USD. Untuk tipe buku, itu bersifat subjektif karena disesuaikan dengan publisher dan konsumen masing masing
